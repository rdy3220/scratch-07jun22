```python
import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot as plt
url = 'https://data.transportation.gov/api/views/85tf-25kj/rows.csv'
data = pd.read_csv(url)
data.head()
```

Questions: 
1) Control for duplicates / data cleaning
2) What accident cause codes have increased in severity (damage cost, injury in '21 vs '19-'20) 
3) Statistical analysis e.g. per miles traveled

```python
'''Remove unwanted columns'''
cols = ['Reporting Railroad Code', 'Reporting Railroad Name', 'Accident Number', 'Date', 'Time', 'Accident Type', 'Persons Evacuated', 
        'State Code', 'County Code', 'Temperature', 'Visibility', 'Weather Condition', 'Train Direction', 'Train Speed', 'Gross Tonnage', 
        'Total Damage Cost', 'Head End Locomotives', 'Mid Train Manual Locomotives', 'Mid Train Remote Locomotives', 'Rear End Manual Locomotives', 
        'Rear End Remote Locomotives', 'Loaded Freight Cars', 'Loaded Passenger Cars', 'Empty Freight Cars', 'Empty Passenger Cars', 
        'Total Persons Killed', 'Total Persons Injured', 'Latitude', 'Longitude']
df = data[cols]

'''Create datetime index'''
df['Date'] = pd.to_datetime(df['Date'])
ix = []
for row, obj in df[['Date', 'Time']].iterrows():
    try: 
        t = dt.datetime.strptime(obj['Time'], '%I:%M %p')
        ix.append(
        (dt.datetime(obj['Date'].year, obj['Date'].month, obj['Date'].day, t.hour, t.minute), int(row))
        )
    except:
            try: 
                ix.append(
                    (dt.datetime(obj['Date'].year, obj['Date'].month, obj['Date'].day, 0, 0), int(row))
                )
            except: 
                ix.append((None, int(row)))
ix = pd.DataFrame(ix, columns=['datetime', 'index']).set_index('index')
df = df.join(ix).set_index('datetime')
df.drop(columns=['Date', 'Time'], inplace=True)
```

```python
'''Set up DataFrames for plotting'''
COL = 'Total Damage Cost'  # replace this with another column e.g. 'Total People Killed'
BASELINE_START = dt.date(2010, 1, 1)  # start of baseline window
BASELINE_END = dt.date(2020, 12, 31)  # end of baseline window
by_month = df[['Accident Type', COL]].pivot_table(index='datetime', values=COL, columns='Accident Type')  # pivot accidents to sum specified COL value over time by accident type
by_month = by_month.groupby(by_month.index.map(lambda x: dt.date(x.year, x.month, 1))).sum()  # bucket into month-year
baseline = by_month.loc[BASELINE_START:BASELINE_END]
baseline_monthly_avg = baseline.groupby(baseline.index.map(lambda x: x.month)).mean()
baseline_monthly_stdev = baseline.groupby(baseline.index.map(lambda x: x.month)).agg(np.std)
baseline_monthly_plus, baseline_monthly_minus = baseline_monthly_avg + baseline_monthly_stdev, baseline_monthly_avg - baseline_monthly_stdev
baseline_monthly_avg_cum = baseline_monthly_avg.cumsum()
baseline_monthly_plus_cum = baseline_monthly_plus.cumsum()
baseline_monthly_minus_cum = baseline_monthly_minus.cumsum()
ytd = by_month.loc[BASELINE_END:dt.date(2021, 12, 31)].cumsum()
ytd.set_index(ytd.index.map(lambda x: x.month), inplace=True)
```

```python
accident_types = ytd.columns
rng = list(np.arange(0, len(accident_types)))

fig, axes = plt.subplots(nrows=len(accident_types), sharex=True, figsize=(5, 5*len(accident_types)))

for i, accident_type in zip(rng, accident_types):
    ax = axes[i]
    ax.set_title(accident_type)
    avg = baseline_monthly_avg_cum[accident_type]
    plus = baseline_monthly_plus_cum[accident_type]
    # minus = baseline_monthly_minus_cum[accident_type]
    actual = ytd[accident_type]
    ax.plot(avg)
    ax.plot(plus)
    # ax.plot(minus)
    ax.plot(actual)
```
